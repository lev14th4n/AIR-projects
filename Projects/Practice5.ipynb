### Visualization with Seaborn

<div style="direction:rtl">
  .

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

<div style="direction:rtl">
  .

```python
sns.set_style('whitegrid')
sns.set_theme(style="white", color_codes=True)
```

<div style="direction:rtl">   
  .

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

<div style="direction:rtl">
          :
<div style="direction:rtl">
:          .

```python
grid = sns.JointGrid(x='fare', y='age', data=titanic)
grid.plot(sns.scatterplot,sns.histplot)
```

<div style="direction:rtl">
         :

```python
titanic.hist("fare",bins=20)
```

<div style="direction:rtl">
          .

```python
sns.boxplot(y='age',x='class',data=titanic)
```

<div style="direction:rtl">
                 .

```python
sns.swarmplot(y='age',x='class',data=titanic)
```

<div style="direction:rtl">
           .

```python
sns.countplot(x='sex',data=titanic)
```

<div style="direction:rtl">
             Titanic Correlations  .

```python
sns.heatmap(titanic.corr())
```

<div style="direction:rtl">
                 .

```python
grid = sns.FacetGrid(data=titanic,col='sex' )
grid.map(plt.hist, 'age')
```
